---

## 1. Conceptual Foundation

In **supervised learning**, the quality of predictions must be assessed relative to the true outcomes.  

- The **Scoring API** in `scikit-learn` formalizes this by mapping predictions $\hat{y}$ (or probability estimates $\hat{p}$) to a **scalar score** representing model performance.  
- Scores can be **maximized** (e.g., accuracy, F1, ROCâ€“AUC) or **minimized** (e.g., mean squared error).  
- To maintain consistency, the API applies a **maximization convention**:  
  - Higher is always better.  
  - Negative signs are applied internally for error functions such as MSE.  

---

## 2. Mechanics of the API

Models can be evaluated via functions such as:

- `cross_val_score`  
- `cross_validate`  
- `GridSearchCV`  
- `RandomizedSearchCV`  

The **scoring parameter** accepts:

1. **Predefined metric names** (e.g., `"accuracy"`, `"f1"`, `"roc_auc"`, `"neg_mean_squared_error"`).  
2. **Custom callables**, enabling domain-specific evaluation criteria.  
3. **Dictionaries of multiple metrics**, allowing simultaneous evaluation across dimensions.  

ðŸ‘‰ This **modular design** ensures **reproducibility** and **comparability** across experiments.

---

## 3. Mathematical Formalization

For a general prediction task, a scoring function can be represented as:

$$
s = S(y, \hat{y})
$$

where:  
- $y$ = ground truth labels or responses  
- $\hat{y}$ = predicted labels or estimated values  
- $S(\cdot)$ = scoring function mapping $(y, \hat{y}) \to \mathbb{R}$  

### Examples

- **Classification Accuracy**  

$$
S(y, \hat{y}) = \frac{1}{n} \sum_{i=1}^n \mathbf{1}(y_i = \hat{y}_i)
$$

- **Mean Squared Error (MSE)**  

$$
S(y, \hat{y}) = -\frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
$$

(Negative sign ensures maximization convention in `scikit-learn`.)

- **ROC AUC**  

$$
S(y, \hat{p}) = \int_0^1 TPR(FPR^{-1}(t)) \, dt
$$

---

## 4. Academic Significance

- The Scoring API encapsulates **best practices** in model evaluation, ensuring metrics are applied **consistently** across tasks.  
- It promotes **comparative research**, enabling algorithms and hyperparameters to be benchmarked under identical scoring schemes.  
- Its **extensibility via custom scorers** aligns with **domain adaptation principles**, where evaluation criteria differ between fields (e.g., medicine vs finance).  

---

## âœ… Summary

The **Scoring API in scikit-learn** provides a **methodologically rigorous interface** for evaluation:  

- Abstracts metrics into a **flexible, reproducible, and standardized framework**.  
- Unifies disparate metrics under a **maximization convention**.  
- Enables **multi-metric evaluation**, advancing both **theoretical clarity** and **practical applicability** in machine learning research.  


# ðŸ“Š Unified Mathematical Framework of Scoring Functions

| **Metric** | **Equation** | **Type** | **Interpretation** |
|------------|--------------|----------|--------------------|
| **Accuracy** | $$ \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN} $$ | Classification | Fraction of correct predictions over total cases. |
| **Precision** | $$ \text{Precision} = \frac{TP}{TP + FP} $$ | Classification | Proportion of predicted positives that are true positives. |
| **Recall (Sensitivity, TPR)** | $$ \text{Recall} = \frac{TP}{TP + FN} $$ | Classification | Fraction of actual positives correctly identified. |
| **Specificity (TNR)** | $$ \text{Specificity} = \frac{TN}{TN + FP} $$ | Classification | Fraction of actual negatives correctly identified. |
| **F1 Score** | $$ F_1 = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}} $$ | Classification | Harmonic mean of precision and recall. |
| **ROC AUC** | $$ AUC = \int_0^1 TPR(FPR^{-1}(t)) \, dt $$ | Classification | Probability that classifier ranks a random positive higher than a random negative. |
| **Log Loss (Cross-Entropy)** | $$ \text{LogLoss} = -\frac{1}{n} \sum_{i=1}^n \big[ y_i \log(\hat{p}_i) + (1-y_i)\log(1-\hat{p}_i) \big] $$ | Classification | Penalizes probabilistic predictions far from the true label. |
| **Mean Squared Error (MSE)** | $$ MSE = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2 $$ | Regression | Average squared deviation between true and predicted values. |
| **Root Mean Squared Error (RMSE)** | $$ RMSE = \sqrt{MSE} $$ | Regression | Standard deviation of prediction errors. |
| **Mean Absolute Error (MAE)** | $$ MAE = \frac{1}{n} \sum_{i=1}^n |y_i - \hat{y}_i| $$ | Regression | Average absolute deviation, robust to outliers. |
| **$R^2$ (Coefficient of Determination)** | $$ R^2 = 1 - \frac{\sum_{i=1}^n (y_i - \hat{y}_i)^2}{\sum_{i=1}^n (y_i - \bar{y})^2} $$ | Regression | Proportion of variance in $y$ explained by the model. |
| **Explained Variance (EVS)** | $$ EVS = 1 - \frac{\text{Var}(y - \hat{y})}{\text{Var}(y)} $$ | Regression | Measures how much of the variance is captured by predictions. |

---

## âœ… Notes

- $TP, TN, FP, FN$ = True Positives, True Negatives, False Positives, False Negatives.  
- $y_i$ = true value; $\hat{y}_i$ = predicted value; $\hat{p}_i$ = predicted probability; $\bar{y}$ = mean of $y$.  
- `scikit-learn` follows a **maximization convention**, so metrics like MSE are internally negated (`neg_mean_squared_error`).

```python
# =====================================================
# Project: Unified Evaluation with Scikit-learn Scoring API
# =====================================================

# ---- Import Libraries ----
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer, load_diabetes
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import (
    confusion_matrix, classification_report, roc_curve, auc, precision_recall_curve,
    mean_squared_error, mean_absolute_error, r2_score
)
```

```python
# =====================================================
# 1. Classification: Breast Cancer Dataset
# =====================================================
data = load_breast_cancer()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

clf_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', LogisticRegression(max_iter=500))
])
clf_pipe.fit(X_train, y_train)
y_pred = clf_pipe.predict(X_test)
y_proba = clf_pipe.predict_proba(X_test)[:, 1]

print("=== Classification Report ===")
print(classification_report(y_test, y_pred, target_names=data.target_names))
```

```python
# Visual 1: Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=data.target_names, yticklabels=data.target_names)
plt.title("Confusion Matrix (Classification)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# Visual 2: ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_proba)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
# Visual 3: Precision-Recall Curve
prec, rec, _ = precision_recall_curve(y_test, y_proba)
plt.plot(rec, prec, label="PR Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.show()
```

```python
# =====================================================
# 2. Regression: Diabetes Dataset
# =====================================================
reg_data = load_diabetes()
X, y = reg_data.data, reg_data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

reg_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('reg', LinearRegression())
])
reg_pipe.fit(X_train, y_train)
y_pred = reg_pipe.predict(X_test)

print("\n=== Regression Metrics ===")
print("MSE:", mean_squared_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
# Visual 4: Residual Plot
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--')
plt.xlabel("True Values")
plt.ylabel("Predicted Values")
plt.title("Residual Plot (Regression)")
plt.show()
```

```python
# Visual 5: Distribution of Residuals
residuals = y_test - y_pred
sns.histplot(residuals, kde=True, color="red")
plt.title("Distribution of Residuals")
plt.xlabel("Error")
plt.show()
```

```python
# =====================================================
# 3. Cross-validation with Multiple Scoring Functions
# =====================================================
scoring = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']
cv_results = cross_validate(clf_pipe, data.data, data.target, cv=5, scoring=scoring, return_train_score=False)
cv_df = pd.DataFrame(cv_results)

print("\n=== Cross-validation Scores (5-Fold) ===")
print(cv_df.mean())
```

```python
# Visual 6: Barplot of CV Scores
cv_means = cv_df[[f'test_{s}' for s in scoring]].mean()
sns.barplot(x=cv_means.index, y=cv_means.values, palette="viridis")
plt.ylabel("Mean CV Score")
plt.title("Cross-Validation Performance Across Metrics")
plt.xticks(rotation=45)
plt.show()
```
