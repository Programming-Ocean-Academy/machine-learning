# Which scoring function should I use?

The choice of a scoring function in machine learning is fundamentally tied to the objective of the predictive task and the statistical properties of the data. Different metrics capture different aspects of model performance, and no single measure is universally optimal.

---

## ðŸ“Œ Classification Tasks

If classes are **balanced**, accuracy is an appropriate measure:

$$
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
$$

For **imbalanced datasets**, accuracy becomes misleading. Alternative functions include:

**Precision & Recall:**

$$
\text{Precision} = \frac{TP}{TP + FP}, \quad
\text{Recall} = \frac{TP}{TP + FN}
$$

**F1-score** (harmonic mean of precision and recall):

$$
F_1 = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
$$

Other threshold-independent metrics include **ROCâ€“AUC** and **PRâ€“AUC**.

---

## ðŸ“Œ Regression Tasks

**Mean Squared Error (MSE)**: penalizes larger errors more heavily.

$$
MSE = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
$$

**Mean Absolute Error (MAE)**: robust to outliers.

$$
MAE = \frac{1}{n} \sum_{i=1}^n \lvert y_i - \hat{y}_i \rvert
$$

**$R^2$ (coefficient of determination)**: measures explained variance.

$$
R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}
$$

---

## ðŸ“Œ Ranking / Probabilistic Predictions

**Log-loss (Cross-Entropy Loss):**

$$
\text{LogLoss} = -\frac{1}{n} \sum_{i=1}^n \sum_{c=1}^C y_{ic} \, \log \hat{p}_{ic}
$$

**Brier Score**: measures calibration of probability estimates.

---

## ðŸ“Œ Concise Guideline

- **Balanced classification** â†’ Accuracy.  
- **Imbalanced classification** â†’ F1, Precisionâ€“Recall AUC.  
- **Regression with outliers** â†’ MAE.  
- **Regression without outliers** â†’ MSE or RMSE.  
- **Probabilistic predictions** â†’ Log-loss, Brier Score.  

---

ðŸ‘‰ Thus, the scoring function must be **aligned with the application context**.  

- In **medical diagnosis** (where false negatives are costly), *Recall* or *F1* is preferable.  
- In **finance** (where probability calibration matters), *Log-loss* or *AUC* is more suitable.

```python
# =====================================================
# Lab: Quantitative Paradigms of Model Evaluation
# =====================================================

# ---- Import Libraries ----
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer, load_diabetes
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score, recall_score, roc_auc_score,
    log_loss, confusion_matrix, classification_report, roc_curve, precision_recall_curve,
    mean_squared_error, mean_absolute_error, r2_score
)

# =====================================================
# 1. Classification Example (Breast Cancer Dataset)
# =====================================================

# Load and split data
data = load_breast_cancer()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Build pipeline and train
clf_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', LogisticRegression(max_iter=500))
])
clf_pipe.fit(X_train, y_train)

# Predictions
y_pred = clf_pipe.predict(X_test)
y_proba = clf_pipe.predict_proba(X_test)[:,1]
```

```python
# Evaluate with multiple metrics
print("=== Classification Metrics ===")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))
print("Log Loss:", log_loss(y_test, y_proba))

print("\n=== Classification Report ===")
print(classification_report(y_test, y_pred))
```

```python
# =====================================================
# 2. Regression Example (Diabetes Dataset)
# =====================================================

# Load and split data
reg_data = load_diabetes()
X, y = reg_data.data, reg_data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Build and train regression pipeline
reg_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('reg', LinearRegression())
])
reg_pipe.fit(X_train, y_train)
y_pred = reg_pipe.predict(X_test)
```

```python
# Evaluate with regression metrics
print("\n=== Regression Metrics ===")
print("MSE:", mean_squared_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

# ðŸ“Š Academic Analysis of Regression Results

---

## 1. Reported Metrics

### Mean Squared Error (MSE)

$$
MSE = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2 = 2821.75
$$

MSE heavily penalizes large deviations, showing how far predictions deviate on average in **squared units**.

---

### Root Mean Squared Error (RMSE)

$$
RMSE = \sqrt{MSE} = 53.12
$$

Expressed in the **same units as the target** (disease progression score), making interpretation more intuitive.

---

### Mean Absolute Error (MAE)

$$
MAE = \frac{1}{n} \sum_{i=1}^n \lvert y_i - \hat{y}_i \rvert = 41.92
$$

Represents the **average absolute deviation**. Lower than RMSE, showing **robustness to outliers**.

---

### $R^2$ Score (Coefficient of Determination)

$$
R^2 = 1 - \frac{\sum_{i=1}^n (y_i - \hat{y}_i)^2}{\sum_{i=1}^n (y_i - \bar{y})^2} = 0.477
$$

Indicates that about **47.7% of the variance** in the diabetes progression outcome is explained by the linear regression model.

---

## 2. Interpretation

- The **moderate $R^2$ (0.477)** shows that while the model captures some variance, it leaves more than half unexplained â†’ highlighting the **complexity of the dataset** and the **limitations of linear regression**.  
- The **gap between RMSE (53.1) and MAE (41.9)** suggests the presence of **outliers**: large errors inflate RMSE more than MAE.  
- In a **medical dataset**, prediction errors of Â±40â€“50 units are **non-trivial**, raising concerns about clinical applicability without additional **feature engineering** or **more expressive models**.

---

## 3. Academic Justification

- **Linear regression** serves as a **baseline benchmark**, but it may **underfit** complex relationships in medical data.  
- Advanced models (**Ridge/Lasso regularization, SVR, tree-based regressors**) could potentially yield **higher $R^2$** and **lower errors**.  
- The value of **multi-metric evaluation** is demonstrated: relying solely on $R^2$ would obscure the **distribution and magnitude** of absolute errors.

---

## âœ… Conclusion

This regression evaluation demonstrates that while **linear regression** provides a starting point, the **Diabetes datasetâ€™s complexity** requires more **sophisticated models** for improved predictive power.  

The **joint use of MSE, RMSE, MAE, and $R^2$** ensures a **comprehensive view** of model performance, aligning with **best academic practices** in regression analysis.

# ðŸ“Š Academic Analysis of the Residual Plot

---

## 1. What the Plot Shows

- **X-axis:** True values of the target (disease progression score).  
- **Y-axis:** Predicted values by the linear regression model.  
- **Dashed diagonal line:** Perfect prediction line ($y = \hat{y}$).  
- **Blue scatter points:** Actual prediction outcomes.  

ðŸ‘‰ Ideally, points should lie close to the diagonal, indicating **accurate predictions**.

---

## 2. Observations

- Many predictions fall **below the diagonal line**, particularly for **higher target values ($>200$)** â†’ systematic **underestimation** for patients with high progression scores.  
- For **lower target values ($<100$)**, predictions fluctuate **both above and below the line**, showing **higher variance** and less reliability for small-value predictions.  
- The scatter is **wide rather than tightly clustered**, reflecting the **moderate $R^2$ value (0.477)** observed earlier â€” the model only **partially explains the variance**.  

---

## 3. Statistical Interpretation

Residuals are defined as:

$$
\epsilon_i = y_i - \hat{y}_i
$$

From the plot:

- $\epsilon_i$ **increases with larger $y_i$**, suggesting **heteroscedasticity** (non-constant variance of residuals).  
- This **violates a key assumption** of classical linear regression.  
- Indicates that a **transformation** (e.g., log scaling of targets) or a **non-linear model** may be more appropriate.

---

## 4. Justification in an Academic Context

- The residual plot supports earlier quantitative findings: relatively **high MAE and RMSE** arise from **systematic underestimation** of extreme values.  
- This aligns with findings in medical prediction literature, where **linear models fail to capture complex, nonlinear relationships** (see Tibshirani, 1996; Hastie, Tibshirani & Friedman, *The Elements of Statistical Learning*).  
- Consequently, while linear regression serves as a **baseline benchmark**, diagnostic evidence suggests that **advanced models** (Ridge/Lasso regression, Random Forests, Gradient Boosting) would better capture the **signal**.  

---

## âœ… Conclusion

The **residual plot** exposes **structural weaknesses** of the linear regression model in predicting diabetes progression:  

- Systematic **underestimation of high scores**.  
- **Heteroscedastic residuals**, violating linear assumptions.  

Thus, a simple linear fit is **inadequate for robust clinical interpretation**, justifying the **academic need for advanced models**.

## 1. Purpose of Cross-Validation

Cross-validation (CV) is a **resampling method** that partitions the dataset into $k$ folds (here $k = 5$) to evaluate **model generalization**.  

- Each fold serves once as the **validation set**, while the remaining folds are used for **training**.  
- The scores are then **averaged**, producing more reliable estimates than a single trainâ€“test split.  

Mathematically, for a metric $M$:

$$
\hat{M}_{CV} = \frac{1}{k} \sum_{i=1}^k M^{(i)}
$$

where $M^{(i)}$ is the score on the $i^{th}$ fold.
