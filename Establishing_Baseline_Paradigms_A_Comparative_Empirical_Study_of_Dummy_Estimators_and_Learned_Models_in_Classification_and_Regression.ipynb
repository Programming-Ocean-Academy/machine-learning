# üìò Dummy Estimators: Baseline Evaluation in Predictive Modeling

---

## üîπ Concept and Purpose

In machine learning, **dummy estimators** are intentionally simplistic models that generate predictions using naive strategies (e.g., random guessing, predicting the most frequent label, or using the mean of the target variable).

- Their primary role is **not to solve the predictive task**, but to serve as a **baseline reference**.  
- This ensures that more complex models (e.g., logistic regression, random forest, neural network) provide **added value beyond trivial heuristics**.  

---

## üîπ Types of Dummy Strategies

### Classification

- **Most Frequent Strategy**

$$
\hat{y} = \arg \max_{c \in C} P(y = c)
$$

Always predicts the majority class.  

- **Stratified Strategy**

$$
P(\hat{y} = c) = \frac{\text{count}(y = c)}{N}
$$

Predicts classes proportionally to their observed frequencies.  

- **Uniform Strategy**

$$
P(\hat{y} = c) = \frac{1}{|C|}
$$

Predicts classes uniformly at random.  

- **Constant Strategy**  
Always predicts a pre-specified label.  

---

### Regression

- **Mean Strategy**

$$
\hat{y} = \frac{1}{N} \sum_{i=1}^N y_i
$$

Predicts the mean of the training targets.  

- **Median Strategy**

$$
\hat{y} = \text{median}(y_1, y_2, \dots, y_N)
$$

Predicts the median of the training targets.  

- **Quantile Strategy**  
Predicts a fixed quantile of the training targets.  

- **Constant Strategy**  
Always predicts a user-defined constant.  

---

## üîπ Mathematical and Statistical Framing

Dummy estimators function as **statistical null models**:

- **Classification baseline accuracy**

$$
Acc_{\text{baseline}} = \max_c \frac{\text{count}(y = c)}{N}
$$

Equivalent to the relative frequency of the majority class.  

- **Regression minimization properties**

The mean predictor minimizes **Mean Squared Error (MSE):**

$$
\hat{y}_{mean} = \arg \min_{\hat{y}} \sum (y_i - \hat{y})^2
$$

The median predictor minimizes **Mean Absolute Error (MAE):**

$$
\hat{y}_{median} = \arg \min_{\hat{y}} \sum |y_i - \hat{y}|
$$

‚û°Ô∏è Thus, dummy estimators are **mathematically optimal** under trivial constraints.  

---

## üîπ Why They Matter

- **Benchmarking** ‚Üí Ensures advanced models surpass trivial baselines.  
- **Sanity Check** ‚Üí Detects data leakage or flawed evaluation (if a complex model underperforms a dummy).  
- **Interpretability** ‚Üí Provides **lower bounds** for expected performance, especially in noisy datasets.  

---

## üîπ Academic Context

- **Dietterich (1998)**: Highlighted the importance of baselines in preventing misleading claims of improvement.  
- **Hastie, Tibshirani & Friedman (2009)**: Discussed mean and median predictors as natural **risk minimizers** under squared and absolute loss.  
- **Scikit-learn Documentation**: Institutionalizes dummy estimators as part of the evaluation API.  

---

## ‚úÖ Verdict

Dummy estimators are not trivial add-ons.  
They are **statistically grounded reference models** that enforce **rigorous empirical validation**, ensuring that machine learning research and applications demonstrate **true predictive value beyond naive baselines**.  


# üìê Mathematical & Statistical Foundations of Dummy Estimators

---

## 1. Dummy Classifiers

### a. Most Frequent Class Baseline
Always predicts the **majority class**:

$$
\hat{y} = \arg \max_{c \in C} P(y = c)
$$

Where:  
- $C$: set of possible classes  
- $P(y = c)$: empirical frequency of class $c$  

---

### b. Stratified Baseline
Predicts classes randomly according to their frequency distribution:

$$
P(\hat{y} = c) = \frac{n_c}{N}
$$

Where:  
- $n_c$: number of samples in class $c$  
- $N$: total number of samples  

---

### c. Uniform Baseline
Predicts classes **uniformly at random**:

$$
P(\hat{y} = c) = \frac{1}{|C|}
$$

---

### d. Evaluation Metrics (Classification)

- **Accuracy**

$$
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
$$

- **Precision**

$$
\text{Precision} = \frac{TP}{TP + FP}
$$

- **Recall (Sensitivity)**

$$
\text{Recall} = \frac{TP}{TP + FN}
$$

- **F1 Score**

$$
F_1 = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
$$

- **ROC AUC**

$$
AUC = \int_0^1 TPR(FPR) \, dFPR
$$

Where:  
- $TP, TN, FP, FN$: true/false positives/negatives  
- $TPR = \frac{TP}{TP+FN}$ (true positive rate)  
- $FPR = \frac{FP}{FP+TN}$ (false positive rate)  

---

## 2. Dummy Regressors

### a. Mean Baseline
Always predicts the **mean** of the training targets:

$$
\hat{y} = \frac{1}{N} \sum_{i=1}^N y_i
$$

---

### b. Median Baseline
Always predicts the **median** of the training targets:

$$
\hat{y} = \text{median}(y_1, y_2, \dots, y_N)
$$

---

### c. Quantile Baseline
Always predicts a specified quantile $q$ of the targets:

$$
\hat{y} = Q_q(y)
$$

---

### d. Evaluation Metrics (Regression)

- **Mean Squared Error (MSE)**

$$
MSE = \frac{1}{N} \sum_{i=1}^N (y_i - \hat{y}_i)^2
$$

- **Root Mean Squared Error (RMSE)**

$$
RMSE = \sqrt{MSE}
$$

- **Mean Absolute Error (MAE)**

$$
MAE = \frac{1}{N} \sum_{i=1}^N |y_i - \hat{y}_i|
$$

- **Coefficient of Determination ($R^2$)**

$$
R^2 = 1 - \frac{\sum_{i=1}^N (y_i - \hat{y}_i)^2}{\sum_{i=1}^N (y_i - \bar{y})^2}
$$

Where:  
- $y_i$: true values  
- $\hat{y}_i$: predicted values  
- $\bar{y}$: mean of true values  

---

## 3. Statistical Justification

Dummy estimators formalize the **null hypothesis**: predictions contain **no learning signal**.  

- **Classification baselines**: majority class prediction or random guessing.  
- **Regression baselines**: mean or median constant predictions.  

‚û°Ô∏è Any valid model must achieve performance **significantly better than these baselines** to demonstrate **true predictive learning**.

## ‚úÖ Academic Justification

Together, these works establish dummy estimators as **statistically grounded null models**:  

- They act as the **minimum performance benchmark**.  
- They ensure **empirical validation** by requiring real models to beat trivial baselines.  
- They connect theory (**no free lunch, risk minimization**) with practice (**scikit-learn Dummy estimators, ROC/Cost curves**).  

```python
# =====================================================
# Project: Baseline Evaluation with Dummy Estimators
# =====================================================

# ---- Import Libraries ----
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer, load_diabetes
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.dummy import DummyClassifier, DummyRegressor
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, r2_score, mean_squared_error

# =====================================================
# 1. Classification: Breast Cancer Dataset
# =====================================================
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Dummy Classifiers
dummy_stratified = DummyClassifier(strategy="stratified", random_state=42)
dummy_mostfreq = DummyClassifier(strategy="most_frequent")

# Real Classifier
clf = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', LogisticRegression(max_iter=500))
])

# Fit and evaluate
models = {
    "Dummy (Most Frequent)": dummy_mostfreq,
    "Dummy (Stratified)": dummy_stratified,
    "Logistic Regression": clf
}

results_clf = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    if name == "Logistic Regression":
        y_proba = model.predict_proba(X_test)[:,1]
        auc = roc_auc_score(y_test, y_proba)
    else:
        auc = np.nan
    results_clf[name] = {
        "Accuracy": accuracy_score(y_test, y_pred),
        "F1": f1_score(y_test, y_pred),
        "ROC AUC": auc
    }

df_clf = pd.DataFrame(results_clf).T
print("=== Classification Results ===")
print(df_clf)
```

```python
# Visualize
df_clf.plot(kind="bar", figsize=(8,5))
plt.title("Classification: Dummy vs Real Model")
plt.ylabel("Score")
plt.xticks(rotation=30)
plt.show()
```

```python
# =====================================================
# 2. Regression: Diabetes Dataset
# =====================================================
X, y = load_diabetes(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Dummy Regressors
dummy_mean = DummyRegressor(strategy="mean")
dummy_median = DummyRegressor(strategy="median")
```

```python
# Real Regressor
reg = Pipeline([
    ('scaler', StandardScaler()),
    ('reg', LinearRegression())
])

# Fit and evaluate
models_reg = {
    "Dummy (Mean)": dummy_mean,
    "Dummy (Median)": dummy_median,
    "Linear Regression": reg
}

results_reg = {}
for name, model in models_reg.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    results_reg[name] = {
        "MSE": mean_squared_error(y_test, y_pred),
        "R2": r2_score(y_test, y_pred)
    }

df_reg = pd.DataFrame(results_reg).T
print("\n=== Regression Results ===")
print(df_reg)
```

```python
# Visualize
fig, axes = plt.subplots(1, 2, figsize=(12,5))
sns.barplot(data=df_reg.reset_index(), x="index", y="MSE", ax=axes[0], palette="coolwarm")
axes[0].set_title("Regression: MSE Comparison")
axes[0].set_ylabel("MSE"); axes[0].set_xlabel("Model")
axes[0].set_xticklabels(df_reg.index, rotation=30)

sns.barplot(data=df_reg.reset_index(), x="index", y="R2", ax=axes[1], palette="viridis")
axes[1].set_title("Regression: R¬≤ Comparison")
axes[1].set_ylabel("R¬≤"); axes[1].set_xlabel("Model")
axes[1].set_xticklabels(df_reg.index, rotation=30)

plt.tight_layout()
plt.show()
```
