## üîπ Abstract  
**Randomized parameter optimization** is a hyperparameter tuning strategy that improves upon exhaustive grid search by sampling a fixed number of random combinations from the hyperparameter space.  

This method offers a more **efficient exploration** of high-dimensional or continuous search spaces, often achieving **comparable or superior results** with significantly reduced computational cost.  

---

## üîπ Methodology  

Let an estimator be parameterized by hyperparameters $\theta \in \Theta$, where $\Theta$ may be large or continuous.  

Instead of evaluating all combinations as in exhaustive grid search, randomized search samples $n$ configurations according to a specified probability distribution:  

$$
\{\theta^{(1)}, \theta^{(2)}, \dots, \theta^{(n)}\} \sim P(\Theta)
$$  

Each sampled $\theta^{(i)}$ is evaluated with **$K$-fold cross-validation**, producing validation scores:  

$$
CV(\theta^{(i)}) = \frac{1}{K} \sum_{k=1}^K M_k(\theta^{(i)})
$$  

where $M_k(\theta^{(i)})$ is the performance metric on fold $k$.  

The optimal hyperparameters are selected as:  

$$
\theta^\ast = \arg\max_{\theta^{(i)}} \; CV(\theta^{(i)})
$$  

---

## üîπ Advantages  

- **Computational Efficiency** ‚Üí Requires fewer evaluations compared to exhaustive search.  
- **Scalability** ‚Üí Suitable for large or continuous hyperparameter spaces.  
- **Flexibility** ‚Üí Can incorporate non-uniform sampling distributions, emphasizing promising regions of the parameter space.  

---

## üîπ Limitations  

- **Stochastic Nature** ‚Üí Results depend on random sampling and may vary across runs.  
- **No Absolute Guarantee** ‚Üí Does not guarantee finding the global optimum, though often sufficient in practice.  

---

## üîπ Conclusion  

Randomized parameter optimization provides a **probabilistic yet efficient alternative** to exhaustive grid search.  

By leveraging random sampling, it balances **exploration** and **computational tractability**, making it particularly effective in scenarios with **high-dimensional** or **continuous** hyperparameter spaces.  

It is therefore regarded as a **practical and widely adopted strategy** in modern machine learning pipelines.  


# üìê Mathematical Equations for Randomized Parameter Optimization

---

## 1. Estimator with Hyperparameters  

A learning algorithm is parameterized by weights $w$ (learned) and hyperparameters $\theta$ (tuned):  

$$
f(x; w, \theta) \;\; \to \;\; \hat{y}
$$  

where $\hat{y}$ is the model prediction.  

---

## 2. Hyperparameter Space  

The hyperparameter search space is defined as:  

$$
\Theta = \Theta_1 \times \Theta_2 \times \cdots \times \Theta_d
$$  

Each $\Theta_j$ corresponds to a hyperparameter domain (continuous, discrete, categorical).  

**Example:**  

- $\Theta_C = [0.01, 10]$  
- $\Theta_\gamma = [0.001, 1]$  
- $\Theta_{\text{kernel}} = \{\text{rbf}, \text{linear}, \text{poly}, \text{sigmoid}\}$  

---

## 3. Random Sampling of Configurations  

Randomized search samples $n$ hyperparameter configurations independently from a probability distribution $P(\Theta)$:  

$$
\{\theta^{(1)}, \theta^{(2)}, \dots, \theta^{(n)}\} \sim P(\Theta)
$$  

- **Continuous domains** ‚Üí e.g., $\theta^{(i)} \sim \text{Uniform}(a,b)$  
- **Discrete domains** ‚Üí e.g., $\theta^{(i)} \sim \text{Categorical}(\{v_1, v_2, \dots, v_m\})$  

---

## 4. Cross-Validation Score for Each Sample  

For each sampled configuration $\theta^{(i)}$, performance is estimated using $K$-fold cross-validation:  

$$
CV(\theta^{(i)}) = \frac{1}{K} \sum_{k=1}^K M_k(\theta^{(i)})
$$  

where $M_k(\theta^{(i)})$ is the performance metric (e.g., accuracy, F1, MSE) on fold $k$.  

---

## 5. Best Parameter Selection  

The optimal parameters are chosen as the maximizer of the CV scores:  

$$
\theta^\ast = \arg\max_{\theta^{(i)} \in \{\theta^{(1)}, \dots, \theta^{(n)}\}} \; CV(\theta^{(i)})
$$  

---

## 6. Evaluation Metrics  

- **Accuracy (Classification):**  

$$
\text{Accuracy} = \frac{1}{N} \sum_{i=1}^N \mathbf{1}\big(\hat{y}^{(i)} = y^{(i)}\big)
$$  

- **Mean Squared Error (Regression):**  

$$
MSE = \frac{1}{N} \sum_{i=1}^N \big(y^{(i)} - \hat{y}^{(i)}\big)^2
$$  

- **F1 Score (Classification):**  

$$
F_1 = \frac{2 \cdot \text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
$$  

---

## 7. Efficiency Advantage  

- **Grid Search:**  

$$
O\left(\prod_{j=1}^d |\Theta_j|\right)
$$  

- **Random Search:**  

$$
O(n), \quad \text{where $n$ is the number of random samples}
$$  

**Bergstra & Bengio (2012):**  
For high-dimensional spaces, random search has a higher probability of sampling near-optimal configurations with far fewer evaluations.  

---

# ‚úÖ Summary  

Randomized optimization formalizes **hyperparameter tuning as a stochastic sampling problem**.  

- Defines **estimator and hyperparameter space**.  
- Uses **random sampling distributions**.  
- Scores each configuration with **cross-validation**.  
- Selects the **best configuration** based on validation performance.  

‚ö° Provides a **computationally efficient alternative** to exhaustive grid search, especially in **high-dimensional** or **continuous** domains.

```python
# Randomized Search with Visualizations (Scikit-learn)

import warnings
warnings.filterwarnings("ignore")
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import uniform
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split, StratifiedKFold, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
import pandas as pd

# Load dataset
digits = load_digits()
X, y = digits.data, digits.target

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Pipeline
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", SVC(probability=True))
])

# Parameter distributions
param_dist = {
    "clf__C": uniform(0.01, 10),
    "clf__gamma": uniform(0.001, 1),
    "clf__kernel": ["rbf", "linear", "poly", "sigmoid"]
}

# Randomized search
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
random_search = RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=param_dist,
    n_iter=30,  # more samples for better visuals
    cv=cv,
    scoring="accuracy",
    random_state=42,
    n_jobs=-1,
    return_train_score=True
)
random_search.fit(X_train, y_train)

# Convert results to DataFrame
results = pd.DataFrame(random_search.cv_results_)
```

```python
plt.figure(figsize=(7,5))
scatter = sns.scatterplot(
    data=results,
    x="param_clf__C",
    y="param_clf__gamma",
    hue="mean_test_score",
    palette="viridis",
    s=100
)
plt.xscale("log")
plt.yscale("log")
plt.title("Randomized Search: Accuracy vs (C, Gamma)")
plt.xlabel("C (log scale)")
plt.ylabel("Gamma (log scale)")

# ‚úÖ seaborn auto-handles legend
plt.legend(title="Mean CV Accuracy", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

```python
### 2. Heatmap of Top Configurations
top_results = results.nlargest(20, "mean_test_score").pivot_table(
    values="mean_test_score",
    index="param_clf__kernel",
    columns="param_clf__C"
)
plt.figure(figsize=(8,5))
sns.heatmap(top_results, annot=True, cmap="magma", fmt=".3f")
plt.title("Top 20 Configurations (Kernel vs C)")
plt.show()
```

```python
### 3. Distribution of CV Scores
plt.figure(figsize=(7,5))
sns.histplot(results["mean_test_score"], bins=10, kde=True, color="skyblue")
plt.title("Distribution of Randomized Search CV Scores")
plt.xlabel("Mean CV Accuracy")
plt.ylabel("Frequency")
plt.show()
```
