# üìê Mathematical Equations for the Lab

---

## 1. Logistic Regression (Classifier in Pipeline)

The model predicts probability using the sigmoid (logistic) function:

$$
\hat{y} = \sigma(z) = \frac{1}{1 + e^{-z}}
$$

where  

$$
z = w^\top x + b
$$

- $x$: input features  
- $w$: weight vector  
- $b$: bias term  

For multiclass, the **softmax** is used:

$$
P(y=k \mid x) = \frac{\exp(w_k^\top x + b_k)}{\sum_{j=1}^K \exp(w_j^\top x + b_j)}
$$

---

## 2. Cross-Validation (General Formula)

For a dataset $D$ split into $K$ folds:

$$
CV_K = \frac{1}{K} \sum_{i=1}^K M_i
$$

where  
- $M_i$: performance metric (e.g., accuracy) on fold $i$  
- $K$: number of folds  

**Special cases:**  
- K-Fold: standard partition into equal-sized folds  
- Stratified K-Fold: preserves class distribution  
- Leave-One-Out (LOO): $K = n$ (number of samples)  

---

## 3. Accuracy (Evaluation Metric)

$$
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
$$

where  
- $TP$: True Positives  
- $TN$: True Negatives  
- $FP$: False Positives  
- $FN$: False Negatives  

---

## 4. Bias‚ÄìVariance Decomposition

Expected generalization error decomposes into bias, variance, and noise:

$$
\mathbb{E}\big[(y - \hat{f}(x))^2\big] =
\underbrace{\big(\mathbb{E}[\hat{f}(x)] - f(x)\big)^2}_{\text{Bias}^2}
+ \underbrace{\mathbb{E}\big[(\hat{f}(x) - \mathbb{E}[\hat{f}(x)])^2\big]}_{\text{Variance}}
+ \sigma^2
$$

where  
- $f(x)$: true function  
- $\hat{f}(x)$: learned predictor  
- $\sigma^2$: irreducible noise  

---

## 5. Learning Curve

Performance as a function of training size $m$:

$$
LC(m) = \frac{1}{K} \sum_{i=1}^K M_i(m)
$$

- $LC(m)$: average validation score when training on $m$ samples  
- Typically shows **bias‚Äìvariance tradeoff**: training error decreases, validation error stabilizes  

---

## 6. Logistic Regression Cost Function

Minimization via cross-entropy loss:

$$
J(w, b) = -\frac{1}{N} \sum_{i=1}^N \Big[ y^{(i)} \log \hat{y}^{(i)} + (1 - y^{(i)}) \log \big(1 - \hat{y}^{(i)}\big) \Big]
$$

where  

$$
\hat{y}^{(i)} = \sigma(w^\top x^{(i)} + b)
$$

---

## 7. Regularization (Used in Logistic Regression $C$ Parameter)

Penalty term in the objective:

$$
J_{\text{reg}}(w,b) = J(w,b) + \lambda \|w\|^2_2
$$

with  

$$
\lambda = \frac{1}{2C}
$$

```python
# End-to-End ML with Cross-Validation Iterators (Scikit-learn)

### 0. Ignore Warnings
import warnings
warnings.filterwarnings("ignore")

### 1. Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import (
    KFold, StratifiedKFold, LeaveOneOut, ShuffleSplit, cross_val_score
)
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

### 2. Load Dataset
data = load_iris()
X, y = data.data, data.target

### 3. Define Pipeline (Scaling + Logistic Regression)
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(max_iter=1000, multi_class="multinomial"))
])

### 4. K-Fold
kf = KFold(n_splits=5, shuffle=True, random_state=42)
scores_kf = cross_val_score(pipeline, X, y, cv=kf, scoring="accuracy")

### 5. StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores_skf = cross_val_score(pipeline, X, y, cv=skf, scoring="accuracy")

### 6. Leave-One-Out
loo = LeaveOneOut()
scores_loo = cross_val_score(pipeline, X, y, cv=loo, scoring="accuracy")

### 7. ShuffleSplit
ss = ShuffleSplit(n_splits=10, test_size=0.3, random_state=42)
scores_ss = cross_val_score(pipeline, X, y, cv=ss, scoring="accuracy")

### 8. Compare Mean Scores
cv_methods = ["KFold", "StratifiedKFold", "LeaveOneOut", "ShuffleSplit"]
mean_scores = [
    scores_kf.mean(),
    scores_skf.mean(),
    scores_loo.mean(),
    scores_ss.mean()
]

plt.figure(figsize=(8,5))
plt.bar(cv_methods, mean_scores, color="skyblue")
plt.title("Cross-Validation Iterators Comparison (Accuracy)")
plt.ylabel("Mean Accuracy")
plt.ylim(0.9, 1.0)
plt.show()
```
