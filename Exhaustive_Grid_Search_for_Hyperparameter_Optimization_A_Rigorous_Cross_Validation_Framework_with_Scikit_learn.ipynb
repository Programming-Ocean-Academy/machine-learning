## üîπ Abstract  
Exhaustive grid search is a **hyperparameter optimization strategy** in which a predefined set of hyperparameter values is systematically explored. Each combination of hyperparameters is evaluated using **cross-validation**, and the configuration yielding the highest validation performance is selected.  

This procedure guarantees that the optimal parameters within the given search space are identified, albeit at a potentially **high computational cost**.  

---

## üîπ Methodology  

Let an estimator $f(x;\theta)$ depend on hyperparameters $\theta \in \Theta$.  
In exhaustive grid search, we define a **discrete parameter grid**:  

$$
\Theta = \{\theta_1, \theta_2, \dots, \theta_m\}
$$  

For each candidate $\theta_j$, the estimator is trained and evaluated via **$K$-fold cross-validation**.  
The mean validation score is:  

$$
CV(\theta_j) = \frac{1}{K} \sum_{k=1}^K M_k(\theta_j)
$$  

where $M_k(\theta_j)$ is the performance metric on fold $k$.  

The optimal parameters are chosen as:  

$$
\theta^\ast = \arg\max_{\theta_j \in \Theta} \, CV(\theta_j)
$$  

---

## üîπ Results & Discussion  

**Advantages:**  
- ‚úÖ Guarantees finding the best-performing hyperparameters within the defined grid.  
- ‚úÖ Simple and parallelizable.  

**Limitations:**  
- ‚ö†Ô∏è Computationally expensive, especially with large grids and high-dimensional parameter spaces.  
- ‚ö†Ô∏è Sensitive to grid resolution:  
  - A **coarse grid** may miss optimal values.  
  - A **fine grid** may become infeasible.  

---

## üîπ Conclusion  

Exhaustive grid search provides a **rigorous and deterministic framework** for hyperparameter optimization.  

Although **computationally demanding**, it remains a **widely used baseline method** due to its:  
- Simplicity  
- Reproducibility  
- Ability to ensure discovery of the best configuration within a predefined search space.  


# üìê Mathematical & Statistical Equations in Hyperparameter Tuning and Grid Search

---

## 1. Model Definition (Estimator)

A machine learning model is parameterized by **weights** $w$ and **hyperparameters** $\theta$:  

$$
f(x; w, \theta) \; \to \; \hat{y}
$$  

- $w$: learned parameters (e.g., regression coefficients).  
- $\theta$: hyperparameters (e.g., regularization constant $C$, kernel parameter $\gamma$).  
- $\hat{y}$: model prediction.  

---

## 2. Loss Functions (Objective to Minimize)

**Logistic Regression (Cross-Entropy Loss):**

$$
J(w,b;\theta) = -\frac{1}{N} \sum_{i=1}^N \Big[ y^{(i)} \log \hat{y}^{(i)} + (1 - y^{(i)}) \log \big(1 - \hat{y}^{(i)}\big) \Big]
$$  

With **L2 regularization** (hyperparameter $C$):  

$$
J_{\text{reg}}(w,b;C) = J(w,b) + \frac{1}{2C} \, \|w\|^2_2
$$  

---

## 3. Performance Metrics (Evaluation Functions)

- **Accuracy (Classification):**  

$$
\text{Accuracy} = \frac{1}{N} \sum_{i=1}^N \mathbf{1}\big(\hat{y}^{(i)} = y^{(i)}\big)
$$  

- **Mean Squared Error (Regression):**  

$$
MSE = \frac{1}{N} \sum_{i=1}^N \big(y^{(i)} - \hat{y}^{(i)}\big)^2
$$  

- **F1 Score (Harmonic Mean of Precision & Recall):**  

$$
F_1 = \frac{2 \cdot \text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
$$  

---

## 4. Cross-Validation Estimation

Given dataset $D = \{(x_i, y_i)\}_{i=1}^N$ split into $K$ folds, performance for hyperparameters $\theta$:  

$$
CV(\theta) = \frac{1}{K} \sum_{k=1}^K M_k(\theta)
$$  

- $M_k(\theta)$: metric score on fold $k$.  
- $CV(\theta)$: average validation performance.  

---

## 5. Grid Search Optimization  

Given hyperparameter search space $\Theta$:  

$$
\theta^\ast = \arg\max_{\theta \in \Theta} \; CV(\theta)
$$  

where $\theta^\ast$ is the best-performing hyperparameter set under cross-validation.  

---

## 6. Bias‚ÄìVariance Tradeoff  

The expected prediction error decomposes into bias, variance, and noise:  

$$
\mathbb{E}[(y - \hat{f}(x))^2] =
\underbrace{\big(\mathbb{E}[\hat{f}(x)] - f(x)\big)^2}_{\text{Bias}^2} +
\underbrace{\mathbb{E}\big[(\hat{f}(x) - \mathbb{E}[\hat{f}(x)])^2\big]}_{\text{Variance}} +
\sigma^2
$$  

---

## 7. Statistical Significance of Model Selection  

Permutation test $p$-value:  

$$
p = \frac{1 + \sum_{j=1}^n \mathbf{1}(CV^{\pi_j} \geq CV)}{n+1}
$$  

- $CV$: observed cross-validation score.  
- $CV^{\pi_j}$: score with permuted labels.  
- $n$: number of permutations.  

---

## 8. Generalization Error Bound (VC Dimension Approximation)  

With probability $1 - \delta$:  

$$
R(f) \leq \hat{R}(f) + \sqrt{\frac{h \big(\log \frac{2N}{h} + 1\big) - \log \frac{\delta}{4}}{N}}
$$  

- $R(f)$: true risk.  
- $\hat{R}(f)$: empirical risk.  
- $h$: VC dimension of the hypothesis class.  
- $N$: number of training samples.  

---

# ‚úÖ Summary  

This field combines:  
- **Optimization equations** (grid search objective, loss functions).  
- **Statistical metrics** (accuracy, F1, MSE).  
- **Validation equations** (cross-validation average, bias‚Äìvariance tradeoff).  
- **Statistical inference** (permutation tests, generalization bounds).  

Together, they form the **mathematical foundation of model selection and hyperparameter tuning** in machine learning.

```python
# End-to-End ML Project: Exhaustive Grid Search (Scikit-learn)

### Import libraries politely
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

### Load dataset politely (Digits dataset for classification)
digits = load_digits()
X, y = digits.data, digits.target

### Split into training and test sets politely
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

### Define pipeline politely (Scaler + SVM classifier)
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", SVC())
])

### Define parameter grid politely for exhaustive search
param_grid = {
    "clf__C": [0.1, 1, 10],
    "clf__gamma": [1, 0.1, 0.01],
    "clf__kernel": ["rbf", "linear"]
}

### Perform exhaustive grid search politely with cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
grid = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=cv,
    scoring="accuracy",
    n_jobs=-1
)
grid.fit(X_train, y_train)
```

```python
### Print best parameters and CV score politely
print("Best Parameters:", grid.best_params_)
print("Best CV Accuracy:", grid.best_score_)
```

```python
### Evaluate best model on test data politely
y_pred = grid.best_estimator_.predict(X_test)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
### Confusion matrix politely
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=digits.target_names,
            yticklabels=digits.target_names)
plt.title("Confusion Matrix (Best Grid Search Model)")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```
