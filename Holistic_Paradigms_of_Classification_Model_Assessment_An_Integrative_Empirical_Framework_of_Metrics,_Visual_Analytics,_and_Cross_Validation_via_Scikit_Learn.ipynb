# üìñ Academic Explanation: Classification Metrics (3.4.4)

## 1. Conceptual Foundation

Classification metrics constitute the **quantitative framework** for assessing the quality of predictive models that assign instances to discrete categories.  

Unlike regression, where prediction errors can be measured on a continuous scale, classification evaluation requires **statistical and probabilistic measures** that capture correctness, discrimination, calibration, and robustness.  

In **scikit-learn**, the classification metrics are formalized under a unified API, ensuring **consistency and reproducibility** in empirical research. They are essential for:  

- Model selection  
- Comparative studies  
- Domain-specific validation (e.g., **medical diagnosis**, **fraud detection**)  

---

## 2. Core Metrics and Their Interpretations

- **Accuracy**: A baseline measure ‚Äî proportion of correct classifications.  
  - Suitable only when classes are balanced.  
  - Formula:  
    $$
    \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
    $$  

- **Precision & Recall**: Complementary metrics.  
  - Precision ‚Üí reliability of positive predictions:  
    $$
    \text{Precision} = \frac{TP}{TP + FP}
    $$  
  - Recall ‚Üí completeness of positive predictions:  
    $$
    \text{Recall} = \frac{TP}{TP + FN}
    $$  

- **F1-score**: Harmonizes precision and recall.  
  $$
  F1 = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
  $$  

- **ROC-AUC**: Evaluates ranking capability independent of thresholds.  
  - Captures area under ROC curve:  
    $$
    AUC = \int_0^1 TPR(FPR) \, d(\text{FPR})
    $$  

- **Log Loss (Cross-Entropy)**: Measures calibration of probabilistic predictions.  
  $$
  \text{LogLoss} = -\frac{1}{N} \sum_{i=1}^{N} \Big[ y_i \log(\hat{p}_i) + (1-y_i)\log(1-\hat{p}_i) \Big]
  $$  

- **MCC (Matthews Correlation Coefficient)**:  
  $$
  MCC = \frac{TP \cdot TN - FP \cdot FN}{\sqrt{(TP+FP)(TP+FN)(TN+FP)(TN+FN)}}
  $$  

- **Cohen‚Äôs Kappa**: Chance-adjusted agreement.  
  $$
  \kappa = \frac{p_o - p_e}{1 - p_e}
  $$  
  where $p_o$ is observed agreement and $p_e$ is expected agreement by chance.  

‚úÖ Together, these metrics address different dimensions of classifier performance, preventing reliance on a single, potentially misleading indicator.  

---

## 3. Mathematical and Statistical Significance

Each metric corresponds to a **statistical optimization criterion**:  

- Mean predictor accuracy $= \frac{\text{correct predictions}}{\text{total predictions}}$.  
- Mean squared error minimizers translate to **probabilistic log-loss** in binary outcomes.  
- ROC and PR curves integrate trade-offs between true positives and false positives, rooted in **statistical decision theory**.  

Formally, these metrics map observed outcomes $y$ and predicted probabilities $\hat{p}$ into scalar values:  

$$
S(y, \hat{p})
$$  

which serve as **objective functions** for model evaluation.  

---

## 4. Academic and Applied Importance

- **Research methodology**:  
  Classification metrics are the foundation of experimental design in AI and ML research, ensuring results are interpretable and comparable across studies.  

- **Applied contexts**:  
  Different fields prioritize different metrics:  
  - **Recall** in cancer detection ($\text{high cost of false negatives}$).  
  - **Precision** in spam filtering ($\text{high cost of false positives}$).  
  This illustrates the **domain adaptation principle** ‚Äî metrics must align with the practical cost of errors.  

- **Scikit-learn‚Äôs role**:  
  Provides implementations that abstract theoretical measures into accessible, reproducible tools, bridging **statistical theory** and **machine learning practice**.  

---

## ‚úÖ Conclusion

Classification metrics are not merely **numerical summaries** ‚Äî they embody the **theoretical principles of statistical decision-making** and act as the **empirical backbone** for machine learning research and application.  

By adopting a **multi-metric evaluation paradigm**, researchers ensure:  

- Methodological rigor  
- Fairness  
- Domain relevance in predictive modeling  


# üìò 3.4.4. Classification Metrics (Scikit-learn Context)

In machine learning, **classification metrics** provide quantitative ways to evaluate how well a model distinguishes between categories.  
Scikit-learn offers a wide range of metrics under its evaluation module, making it possible to assess models **beyond raw accuracy**.

---

## üîë Core Metrics and Their Mathematical Formulation

### Accuracy
Proportion of correct predictions among all predictions:

$$
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
$$

---

### Precision
Fraction of predicted positives that are truly positive:

$$
\text{Precision} = \frac{TP}{TP + FP}
$$

---

### Recall (Sensitivity / True Positive Rate)
Fraction of actual positives that are correctly identified:

$$
\text{Recall} = \frac{TP}{TP + FN}
$$

---

### F1-Score
Harmonic mean of precision and recall:

$$
F_1 = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
$$

---

### Specificity (True Negative Rate)
Fraction of actual negatives that are correctly identified:

$$
\text{Specificity} = \frac{TN}{TN + FP}
$$

---

### ROC-AUC (Receiver Operating Characteristic ‚Äì Area Under Curve)
Probability that a randomly chosen positive is ranked higher than a randomly chosen negative:

$$
AUC = \int_0^1 TPR(FPR) \, d(FPR)
$$

Where:  
- $TPR = \frac{TP}{TP+FN}$ (True Positive Rate)  
- $FPR = \frac{FP}{FP+TN}$ (False Positive Rate)  

---

### Log-Loss (Cross-Entropy Loss)
Used for **probabilistic classifiers**, penalizing false confidence:

$$
\text{LogLoss} = -\frac{1}{N} \sum_{i=1}^N \big[ y_i \log(\hat{p}_i) + (1-y_i)\log(1-\hat{p}_i) \big]
$$

---

### Cohen‚Äôs Kappa
Measures agreement between predictions and labels while accounting for chance:

$$
\kappa = \frac{p_o - p_e}{1 - p_e}
$$

Where:  
- $p_o$: observed agreement  
- $p_e$: expected agreement by chance  

---

### Matthews Correlation Coefficient (MCC)
Balanced correlation coefficient, even with imbalanced classes:

$$
MCC = \frac{TP \cdot TN - FP \cdot FN}{\sqrt{(TP+FP)(TP+FN)(TN+FP)(TN+FN)}}
$$

## ‚úÖ Implementation in Scikit-learn

- All these metrics are implemented in **Scikit-learn‚Äôs evaluation API**.  
- They can be passed directly into evaluation utilities such as:  
  - `cross_val_score`  
  - `cross_validate`  
  - `GridSearchCV`  

‚û°Ô∏è This standardization ensures **reproducibility**, **comparability**, and the ability to integrate **custom scorers** tailored to domain-specific needs.  


```python
# =====================================================
# Lab: Comprehensive Study of Classification Metrics
# =====================================================

# ---- Import Libraries ----
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, log_loss, confusion_matrix, roc_curve, precision_recall_curve,
    classification_report
)
```

```python
# =====================================================
# 1. Load Data and Build Model
# =====================================================
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

clf = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', LogisticRegression(max_iter=500))
])
clf.fit(X_train, y_train)

# Predictions
y_pred = clf.predict(X_test)
y_proba = clf.predict_proba(X_test)[:,1]
```

```python
# =====================================================
# 2. Numerical Evaluation Metrics
# =====================================================
metrics = {
    "Accuracy": accuracy_score(y_test, y_pred),
    "Precision": precision_score(y_test, y_pred),
    "Recall": recall_score(y_test, y_pred),
    "F1 Score": f1_score(y_test, y_pred),
    "ROC AUC": roc_auc_score(y_test, y_proba),
    "Log Loss": log_loss(y_test, y_proba)
}
print("=== Classification Metrics ===")
for k,v in metrics.items():
    print(f"{k}: {v:.4f}")
```

```python
# =====================================================
# 3. Visualizations
# =====================================================

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Benign","Malignant"], yticklabels=["Benign","Malignant"])
plt.title("Confusion Matrix")
plt.xlabel("Predicted"); plt.ylabel("Actual")
plt.show()
```

```python
# ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_proba)
plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc_score(y_test,y_proba):.3f}")
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
# Precision-Recall Curve
prec, rec, _ = precision_recall_curve(y_test, y_proba)
plt.figure(figsize=(6,5))
plt.plot(rec, prec, label="PR Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.show()
```

```python
# Classification Report Heatmap
report = classification_report(y_test, y_pred, output_dict=True)
df_report = pd.DataFrame(report).iloc[:-1, :].T
plt.figure(figsize=(7,4))
sns.heatmap(df_report, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Classification Report (Precision, Recall, F1, Support)")
plt.show()
```

```python
# =====================================================
# 4. Cross-Validation with Multiple Metrics
# =====================================================
scoring = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']
cv_results = cross_validate(clf, X, y, cv=5, scoring=scoring)
cv_df = pd.DataFrame(cv_results)[[f'test_{s}' for s in scoring]]

print("\n=== Cross-validation Results ===")
print(cv_df.mean())
```

```python
# Visualize CV Results
cv_means = cv_df.mean()
plt.figure(figsize=(7,5))
sns.barplot(x=cv_means.index, y=cv_means.values, palette="viridis")
plt.ylabel("Mean CV Score")
plt.title("Cross-Validation Performance Across Metrics")
plt.xticks(rotation=45)
plt.show()
```
